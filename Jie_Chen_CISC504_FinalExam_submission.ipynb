```python
# Jie Chen CISC504 Final Exam Exercise 2
def main():
    PHRASE = r'The octopus snagged the blue fish'
    dict_element = {'subject': 'octopus', 'verb': 'snagged', 'adjective':'blue', 'object':'fish'}
    print('Original Pharase:')
    printPhrase(dict_element)
    # read user input
    dict_element['subject'] = readString('subject')
    dict_element['verb'] = readString('verb')
    dict_element['adjective'] = readString('adjective')
    dict_element['object'] = readString('object')
    # print new phrase
    print('After replacement:')
    printPhrase(dict_element)

def printPhrase(dict_e):
    print('The ' + dict_e['subject'] +' '+ dict_e['verb'] + ' the ' + dict_e['adjective']  + ' ' + dict_e['object']+'.')

def readString(label):
    while (True):
        s = input('Please enter ' + label + ':\n')       
        if (isinstance(s, str) and len(s) > 0 and s[0].isalpha()):
            return s
        else:
            print('Invalid input of ' + label + " as " + s)

main()
```

```python
# Jie Chen CISC504 Final Exam Exercise 3
import random

def main():
    print('Guessing Number!')
    print('Please pick a random number in the range of [1:100] and remember by your heart.')
    low = 1
    high = 100
    mid = (low + high) // 2
    guess(mid)

def guess(mid):   
    while (True):
        print('Is it ', mid, '?')
        u = input('(Please input Y for "Yes", H for "too high" and L for "too low") :')
        if u == 'Y':
            print(mid, ", I got it! Thank you!")
            break
        elif u == 'H':
            mid -= 1
        elif u == 'L':
            mid += 1
        else:
            print('Invalid input, try again!')
            continue

main()
```
